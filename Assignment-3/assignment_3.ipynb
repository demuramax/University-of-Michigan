```python
import pandas as pd
import numpy as np
import os
import re
```

```python
Energy = pd.read_excel('Energy Indicators.xls', header=16, skipfooter=45)
Energy.head(2)
```

```python
Energy = Energy.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)
Energy.head()
```

```python
Energy.columns
```

```python
Energy = Energy.rename(columns={'Unnamed: 2': 'Country', 'Energy Supply per capita': 'Energy Supply per Capita', 
                        'Renewable Electricity Production': '% Renewable'})
Energy = Energy.drop([Energy.index[0]])
Energy['Country'] = Energy['Country'].apply(lambda x: re.sub(r'\s\(.+\)|\d+', '', x))
Energy['Country'] = Energy['Country'].replace({"Republic of Korea": "South Korea", "United States of America": "United States", 
                                       "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", 
                                       "China, Hong Kong Special Administrative Region": "Hong Kong"})
Energy = Energy.reset_index().set_index(['Country']).drop('index', axis=1)
```

```python
Energy.head()
```

```python
Energy['Energy Supply'].unique()
```

```python
Energy['Energy Supply'] = Energy['Energy Supply'].replace('...',np.nan)
Energy['Energy Supply'].astype('Int64')
```

```python
Energy['Energy Supply'] = Energy['Energy Supply'].apply(lambda x: x*1000000)
```

```python
[x for x in Energy.index]
```

```python
gdp = pd.read_csv('world_bank.csv', header=4)
gdp.head(2)
```

```python
gdp = gdp.drop(gdp.columns[1:50], axis=1)
gdp['Country Name'] = gdp['Country Name'].replace({"Korea, Rep.": "South Korea", 
"Iran, Islamic Rep.": "Iran",
"Hong Kong SAR, China": "Hong Kong"})
```

```python
ScimEn = pd.read_excel('scimagojr-3.xlsx')
ScimEn.head(2)
```

```python
ScimEn = ScimEn[ScimEn['Rank']<=15]
merged = ScimEn.merge(Energy, left_on='Country', right_on=Energy.index)
merged.head(3)
```

```python
merged_df = merged.merge(gdp, left_on='Country', right_on='Country Name').drop('Country Name', axis=1)
merged_df.head(3)
```

```python
merged_df = merged_df.reset_index().set_index(['Country']).drop('index', axis=1)
merged_df
```
